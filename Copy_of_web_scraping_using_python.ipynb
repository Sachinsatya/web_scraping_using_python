```python
import requests
```

```python
wiki_link = 'https://en.wikipedia.org/wiki/Asia#:~:text=They%20are%20Bangladesh%2C%20China%2C%20India,Hong%20Kong%2C%20Tokyo%20and%20Singapore.'
```

```python
link = requests.get(wiki_link).text
```

```python
print(link)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(link, 'lxml')
print(soup)
```

```python
print(soup.prettify())
```

```python
soup.title
```

```python
soup.title.string
```

```python
soup.a
```

```python
soup.find_all('a')
```

```python
all_link = soup.find_all('a')
for link in all_link:
  print(link.get('href'))
```

```python
all_tables = soup.find_all('table')
print(all_tables)
```

```python
right_table = soup.find('table', class_='wikitable')
right_table
```

```python
table_links = right_table.findAll('a')
table_links
```

```python
country = []
for links in table_links:
  country.append(links.get('title'))
print(country)
```

```python
import pandas as pd
df = pd.DataFrame()
df['country'] = country
df
```
